```python
# ============================================================
# ADALINE (Adaptive Linear Neuron) Implementation from Scratch
# ============================================================

# Import NumPy library
# NumPy is used for array handling, mathematical operations,
# dot products, and numerical computations
import numpy as np


# ============================================================
# Define the ADALINE Class
# ============================================================
class Adaline:

    # --------------------------------------------------------
    # Constructor Method
    # --------------------------------------------------------
    def __init__(self, learning_rate=0.01, epochs=50):
        """
        learning_rate : Controls the step size in gradient descent
        epochs        : Number of times the training data is iterated
        """

        # Store learning rate as a class variable
        self.learning_rate = learning_rate

        # Store number of epochs
        self.epochs = epochs


    # --------------------------------------------------------
    # Fit Method (Training the ADALINE Model)
    # --------------------------------------------------------
    def fit(self, X, y):
        """
        X : Input feature matrix (n_samples × n_features)
        y : Target output vector
        """

        # Get number of samples and number of features from X
        n_samples, n_features = X.shape

        # Initialize weights to zero for each feature
        # One weight corresponds to one input feature
        self.weights = np.zeros(n_features)

        # Initialize bias term to zero
        self.bias = 0

        # Initialize a list to store loss values for each epoch
        self.losses = []

        # ---------------- Training Loop ----------------
        for epoch in range(self.epochs):

            # Compute linear combination (net input)
            # net_input = X · w + b
            linear_output = np.dot(X, self.weights) + self.bias

            # Apply linear activation function (identity function)
            # ADALINE uses a linear activation
            y_predicted = linear_output

            # Compute error (difference between actual and predicted output)
            errors = y - y_predicted

            # Update weights using gradient descent rule
            # w = w + η * (Xᵀ · error) / n_samples
            self.weights += self.learning_rate * np.dot(X.T, errors) / n_samples

            # Update bias using gradient descent
            # b = b + η * sum(error) / n_samples
            self.bias += self.learning_rate * np.sum(errors) / n_samples

            # Compute Mean Squared Error (MSE)
            # Loss function minimized by ADALINE
            loss = (errors ** 2).mean() / 2

            # Store loss for this epoch
            self.losses.append(loss)

            # Print loss value for each epoch (optional for observation)
            print(f"Epoch {epoch + 1}/{self.epochs}, Loss: {loss:.4f}")


    # --------------------------------------------------------
    # Predict Method
    # --------------------------------------------------------
    def predict(self, X):
        """
        Predict class labels for input data
        """

        # Compute linear output using trained weights and bias
        linear_output = np.dot(X, self.weights) + self.bias

        # Convert continuous output into class labels
        # If output >= 0 → class +1
        # Else → class -1
        return np.where(linear_output >= 0, 1, -1)


# ============================================================
# Main Program (Testing ADALINE)
# ============================================================

# Define a simple linearly separable dataset
# Each row represents one data point with two features
X = np.array([
    [ 1,  1],
    [ 1, -1],
    [-1,  1],
    [-1, -1]
])

# Define corresponding target labels
# Classes are represented as +1 and -1
y = np.array([1, -1, -1, -1])


# Create an object of the Adaline class
# learning_rate controls how fast the model learns
# epochs define number of training iterations
model = Adaline(learning_rate=0.01, epochs=50)


# Train the ADALINE model using the fit method
model.fit(X, y)


# Predict class labels for the training data
predictions = model.predict(X)


# Display predicted and actual labels
print("\nPredicted Labels:", predictions)
print("Actual Labels   :", y)


# Display final weights and bias
print("\nFinal Weights:", model.weights)
print("Final Bias   :", model.bias)
```
